```python
from embedder import Embedder

embed = Embedder()
```

```python
from tqdm.auto import tqdm
```

Question 1

```python
q1 = 'How does approximate nearest neighbor search work?'

v1 = embed.encode(q1)

v1[0]
```

```python
from gitsource import GithubRepositoryDataReader

reader = GithubRepositoryDataReader(
    repo_owner="DataTalksClub",
    repo_name="llm-zoomcamp",
    commit_id="8c1834d",
    allowed_extensions={"md"},
    filename_filter=lambda path: "/lessons/" in path,
)

documents = [file.parse() for file in reader.read()]
```

Question 2

```python
d = ''
for i in documents:
    if i['filename'] == '02-vector-search/lessons/07-sqlitesearch-vector.md':

        d = i['content']

dv = embed.encode(d)
```

```python
v1.dot(dv)
```

Question 3


```python
from gitsource import chunk_documents
chunks = chunk_documents(documents, size=2000, step=1000)
```

```python
texts = [i['filename'] + ' ' + i['content'] for i in chunks]
```

```python
import numpy as np

vectors = []
batch_size = 50

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]
    batch_vector = embed.encode_batch(batch)
    vectors.extend(batch_vector)

X = np.array(vectors)
```

```python
scores = X.dot(v1)
idx = np.argmax(scores)
chunks[idx]['filename']
```

Question 4 

```python
from minsearch import VectorSearch

vector_index = VectorSearch()
vector_index.fit(X, chunks)
query = 'What metric do we use to evaluate a search engine?'
vectory_query = embed.encode(query)

result = vector_index.search(
    vectory_query,
    num_results=5
)

for i in result:
    print(i['filename'])
```

Question 5

```python
from minsearch import Index
create_index = Index(
    text_fields=['content', 'filename']
)
create_index.fit(chunks)

query = 'How do I store vectors in PostgreSQL?'
result = create_index.search(
    query,
    boost_dict={'content':3.0},
    num_results=5
)

for i in result:
    print(i['filename'])
```
